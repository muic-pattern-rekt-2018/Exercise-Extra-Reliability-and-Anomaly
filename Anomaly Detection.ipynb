# Basic (1D)

```python
import numpy as np
from matplotlib import pyplot as plt
```

1) Find anomaly here.

```python
data = np.random.randn(1000)*3 + 10
data = np.concatenate((data, [24.2]))

plt.hist(data, bins=20)
```

print(2) What is the probability of finding such a data to appear?

```python
print(np.average(data))
print(np.std(data))
```

```python
z = (24.2 - np.average(data))/np.std(data)
print(z)
```

```python
import scipy.stats

p_values = scipy.stats.norm.sf(abs(z))
print(p_values)
```

# Multiple Dimension

```python
xs = np.random.randn(100)*2+3
ys = 2*xs + 3 + np.random.randn(100)

xs = np.concatenate([xs,[4]])
ys = np.concatenate([ys,[0]])

plt.hist2d(xs, ys, bins=20);
```

1) Find the anomaly here. Whats the idea?

```python
plt.hist(xs)
```

```python
plt.hist(ys)
```

```python
w = np.polyfit(xs, ys, 1)



plt.plot(xs, ys, '.');

xxs= np.linspace(-3, 9, 100)
yys = np.polyval(w, xxs)
plt.plot(xxs, yys)
```

```python
yys = np.polyval(w, xs)
plt.plot(ys, yys, '.')
plt.xlabel('actual')
plt.ylabel('prediction')

s = np.linspace(-5, 20, 100)
plt.plot(s, s)
plt.grid()
```

```python
res = yys - ys # pred - actual

res_mean = np.average(res)
res_std = np.std(res)
z_score = np.abs((res - res_mean)/res_std)

plt.hist(res, bins=30)

```

```python

plt.scatter(xs, ys, c=z_score);
plt.colorbar()
```
